## **2.1. Анализ жанров: структура жанров и количество упоминаний в фильмах**

Всего в базе данных 22 уникальных жанра, их общее количество с повторами составляет 1456 (поскольку в одном фильме может быть упомянуто несколько жанров). Наиболее популярными являются:
1. Драма (встречается в 250 фильмах из 477) - 17.2% доля среди всех жанровых вхождений
2. Боевик (164) - 11.3%
3. Комедия (154) - 10.6%
4. Фантастика (115) - 7.9%
5. Мелодрама (109) - 7.5%
6. Криминал (101) - 6.9%
7. Приключения (101) - 6.9%
8. Трилер (97) - 6.7%
9. Фэнтези (83) - 5.7%
10. Ужасы (59) - 4.1%

```python
genre_count = {}

for genres in df['Genres']:
    if pd.isna(genres):
        continue
    for genre in genres.split(','):
        genre = genre.strip()
        if genre in genre_count:
            genre_count[genre] += 1
        else:
            genre_count[genre] = 1
```

```python
# Анализ популярности жанров по кол-ву фильмов
genre_df = pd.DataFrame(list(genre_count.items()), columns=['Genre', 'Count']).sort_values('Count', ascending=False)
total_films = len(df)
genre_df['Percent'] = genre_df['Count'] / total_films * 100 # в Х% фильмов присутствует данный жанр
genre_df
```

```python
# Кол-во уникальных жанров
genre_df['Genre'].nunique()
```

```python
# Общее кол-во указанных жанров (с повторами)
print(genre_df['Count'].sum())
```

```python
# Анализ популярности жанров по кол-ву фильмов
genre_df.plot(x='Genre', y='Count', xlabel = 'Жанр', ylabel = 'Количество', kind='bar', figsize=(12, 5), title='Популярность жанров по количеству фильмов', rot=45)
```

```python
# Анализ структуры жанров
top_n = 10
top = genre_df.head(top_n)
other = pd.DataFrame({'Genre': ['Прочее'], 'Count': [genre_df['Count'][top_n:].sum()]})

genre_pie = pd.concat([top, other]) # объединение таблиц

genre_pie.plot(y='Count', kind='pie', labels=genre_pie['Genre'], figsize=(8, 8), autopct='%1.1f%%', title='Структура жанров: ТОП-10').set_ylabel('')
```

```python
ratings_df = df[['Title', 'Year', 'Genres', 'Country', 'Rating Kinopoisk', 'Rating Imdb']].dropna() # убираем строки (фильмы), где нет данных (не указаны рейтинги)
```

```python
# 0) Найдем разницу между рейтингом фильмов на разных платформах
ratings_df['Rating_Diff'] = (ratings_df['Rating Kinopoisk'] - ratings_df['Rating Imdb'])
ratings_df['Abs_Diff'] = ratings_df['Rating_Diff'].abs()
ratings_df
```

```python
# 1) Фильмы с наибольшим расхождением рейтингов
ratings_df_sorted = ratings_df.sort_values('Abs_Diff', ascending = False)
ratings_df_sorted[['Title', 'Year', 'Genres', 'Country', 'Rating Kinopoisk',	'Rating Imdb', 'Abs_Diff']]
```

```python
# 2) На каких платформах оценка фильмов выше
ratings_df['Rating_Diff'].apply(lambda x: 'Кинопоиск выше' if x > 0 else 'IMDb выше').value_counts()
```

```python
# проверка, связано ли это со страной производства фильма
ratings_df['Country_List'] = ratings_df['Country'].apply(lambda x: [c.strip() for c in x.split(',')])
ratings_df_exploded = ratings_df.explode('Country_List') # один фильм на 1 строке -> несколько строк (по странам) с одним и тем же фильмом
ratings_df_exploded[['Title', 'Country_List']]
```

```python
# cчитаем количество фильмов по странам: оценка на *Кинопоиске* выше

kp_higher = ratings_df_exploded[ratings_df_exploded['Rating_Diff'] > 0]

kp_country_count = kp_higher['Country_List'].value_counts()
kp_country_count.columns = ['Country', 'Films_Count']
kp_country_count.head(6)
```

```python
# cчитаем количество фильмов по странам: оценка на *IMDb* выше

imdb_higher = ratings_df_exploded[ratings_df_exploded['Rating_Diff'] < 0]

imdb_country_count = imdb_higher['Country_List'].value_counts()
imdb_country_count.columns = ['Country', 'Films_Count']
imdb_country_count.head(6)
```

```python
# берём ТОП-6 стран в каждом случае
kp_top = kp_country_count.head(6)
imdb_top = imdb_country_count.head(6)

# приводим к DataFrame
kp_df = kp_top.reset_index()
kp_df.columns = ['Country', 'Кинопоиск выше']
imdb_df = imdb_top.reset_index()
imdb_df.columns = ['Country', 'IMDb выше']

# объединяем по странам (появляются пропуски -> их заполняем 0)
countries_compare = pd.merge(kp_df, imdb_df, on='Country', how='outer').fillna(0) # указан "outer" => ВСЕ страны из обеих таблиц / без него по умолчанию "inner" => страны, которые есть в ОБЕИХ таблицах.

countries_compare
```

```python
countries_compare.sort_values('Кинопоиск выше', ascending = False).plot(kind='bar', x='Country', y=['Кинопоиск выше', 'IMDb выше'], figsize=(10, 5), title='Страны производства фильмов с более высокими рейтингами\nна Кинопоиске и IMDb')
```
